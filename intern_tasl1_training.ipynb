```python
df['Text'] = df['Text'].apply(lambda x: ' '.join(x.lower() for x in x.split()))
```

```python
df['Text'] = df['Text'].str.replace('[^\w\s]',"")
```

```python
from nltk.corpus import stopwords
from textblob import TextBlob
from textblob import Word
```

```python
stop = stopwords.words('english')
df['Text'] = df['Text'].apply(lambda x: ' '.join(x for x in x.split() if x not in stop))
```

```python
df
```

```python
review_1 = df[df['Score']==1]
review_2 = df[df['Score']==2]
review_3 = df[df['Score']==3]
review_4 = df[df['Score']==4]
review_5 = df[df['Score']==5]
```

```python
review_sample = pd.concat([review_1,review_2,review_3,review_4,review_5])
```

```python
review_sample.reset_index(drop = True,inplace= True)
```

```python
review_sample.drop('index',inplace=True,axis=1)
```
